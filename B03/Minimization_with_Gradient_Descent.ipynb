```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2 + 6*np.sin(x)

def f_prime(x):
    return 2*x + 6*np.cos(x)
```

```python
def gradient_descent(x0 = 0, learning_rate = 0.001, max_iter = 1000000, epsilon = 1e-3):
    x = [x0]
    
    for i in range(max_iter):
        x_new = x[-1] - learning_rate*f_prime(x[-1])
        if abs(f_prime(x_new)) <= epsilon:
            break
        x.append(x_new)
    return x 
```

```python
x0 = 0; learning_rate = 0.001; max_iter = 1000000; epsilon = 1e-3
x = gradient_descent(x0, learning_rate, max_iter, epsilon)
```

```python
df = pd.DataFrame({
    'Iteration': range(len(x)),
    'x': x,
    'f(x)': [f(val) for val in x]
})

pd.set_option('display.max_rows', 20)
print(df)

x = np.linspace(-4, 4, 100)
y = f(x)
plt.grid()
plt.plot(x, y)
plt.scatter(x, f(x), c='r')
plt.show()

print(f'Local minimum occurs at x = {x[-1]}')
print(f'Number of iterations: {i}')
print(f'f(x) = {f(x[-1])}')
print(f'f\'(x) = {f_prime(x[-1])}')
```
